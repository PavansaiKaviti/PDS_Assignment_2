```python
def extract_numeric(text):
    if isinstance(text, str):
        numeric_part = ''.join([c for c in text if c.isdigit() or c == '.'])
        return float(numeric_part) if numeric_part else None
    elif isinstance(text, float):
        return text
    else:
        return None
```

```python
data['Power'] = data['Power'].apply(extract_numeric)
```

```python
data['Mileage'] = data['Mileage'].apply(extract_numeric)
```

```python
data['New_Price'] = data['New_Price'].apply(extract_numeric)
```

```python
data=data.fillna(data.median())
```
